```python
import pandas as pd
```

```python
df = pd.read_csv('tripdata.csv')
df
```

```python
missing_percent = df.isnull().mean() * 100
print(missing_percent.round(2))
```

```python
df['start_station_name'] = df['start_station_name'].fillna('Unknown')
df['start_station_id']   = df['start_station_id'].fillna('Unknown')

df['end_station_name']   = df['end_station_name'].fillna('Unknown')
df['end_station_id']     = df['end_station_id'].fillna('Unknown')
```

```python
df = df.dropna(subset=['end_lat', 'end_lng'])
df
```

```python
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
```

```python
df['ride_duration_minutes'] = ((df['ended_at'] - df['started_at']).dt.total_seconds() / 60).round(2)
# Remove negative or zero duration rides
df = df[df['ride_duration_minutes'] > 0]
```

```python
df = df[df['ride_duration_minutes'] <= 1440]  # 1440 minutes = 24 hours
df
```

```python
df['day_of_week'] = df['started_at'].dt.day_name()
df['month'] = df['started_at'].dt.month_name()
df['start_hour'] = df['started_at'].dt.hour
```

```python
# Do casual riders ride longer than members?
avg_duration = (
    df.groupby('member_casual')['ride_duration_minutes']
    .mean()
    .reset_index(name='avg_ride_duration_minutes')
)

# Average ride duration by member type and day of week
avg_duration_by_day = (
    df.groupby(['member_casual', 'day_of_week'])['ride_duration_minutes']
    .mean()
    .reset_index(name='avg_ride_duration_minutes')
)

# Ride count by member type and day of week
ride_count_by_day = (
    df.groupby(['member_casual', 'day_of_week'])['ride_id']
    .count()
    .reset_index(name='ride_count')
)

print(avg_duration.round(2))
print(avg_duration_by_day.round(2))
print(ride_count_by_day.round(2))
```

```python
# Are members more active on weekdays or weekends?
rides_by_day = df.groupby(['member_casual', 'day_of_week'])['ride_id'].count().reset_index(name='ride_count')
rides_by_day
```

```python
# Which months have more casual rider activity?
rides_by_month = df.groupby(['member_casual', 'month'])['ride_id'].count().reset_index(name='ride_count')
rides_by_month
```

```python
#Which stations are most popular for each group?
top_stations = df.groupby(['member_casual', 'start_station_name']) ['ride_id'].count().reset_index(name='ride_count')
top10_stations = top_stations.sort_values(['member_casual', 'ride_count'], ascending=[True, False]).groupby('member_casual').head(10)
top10_stations
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.interpolate import make_interp_spline
```

```python
sns.set_theme(style="whitegrid", font_scale=1.1)

# --- 1. Average Ride Duration by Member Type ---
avg_duration = df.groupby('member_casual')['ride_duration_minutes'].mean().reset_index()

plt.figure(figsize=(6, 4))
sns.barplot(data=avg_duration, x='member_casual', y='ride_duration_minutes', 
            palette=['#007bff', '#0040ff'],width=0.4)
plt.title('Average Ride Duration by Member Type', fontsize=14, weight='bold')
plt.ylabel('Average Duration (minutes)')
plt.xlabel('Member Type')

# Remove grid and borders
plt.grid(False)  
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```

```python
# Keep only Jan to July
valid_months = ['January', 'February', 'March', 'April', 'May', 'June', 'July']
rides_by_month = rides_by_month[rides_by_month['month'].isin(valid_months)]

plt.figure(figsize=(10, 5))
sns.lineplot(data=rides_by_month, x='month', y='ride_count', hue='member_casual',
             marker='o', palette=['#007bff', '#0040ff'])
plt.title('Number of Rides per Month by User Type (Janâ€“July 2025)', fontsize=14, weight='bold')
plt.ylabel('Ride Count')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- 3. Ride Frequency Heatmap by Day & Hour ---
day_order = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
df['day_of_week'] = pd.Categorical(df['day_of_week'], categories=day_order, ordered=True)
rides_by_hour = df.groupby(['day_of_week', df['started_at'].dt.hour])['ride_id'].count().reset_index(name='ride_count')

pivot_heatmap = rides_by_hour.pivot(index='day_of_week', columns='started_at', values='ride_count')

plt.figure(figsize=(12, 6))
sns.heatmap(pivot_heatmap, cmap='Blues', linewidths=0.5)
plt.title('Ride Frequency Heatmap by Day & Hour', fontsize=14, weight='bold')
plt.ylabel('Day of Week')
plt.xlabel('Hour of Day')
plt.tight_layout()
plt.show()
```

```python
sns.set_theme(style="whitegrid", font_scale=1.0)
plt.figure(figsize=(14, 8))
ax = sns.barplot(data=top10_stations, x='start_station_name', y='ride_count', hue='member_casual',
                palette=['#007bff', '#0040ff'], dodge=True)

plt.title('Top 10 Start Stations by User Type', fontsize=16, weight='bold', pad=20)
plt.xlabel('Start Station', fontsize=12, weight='bold')
plt.ylabel('Number of Rides', fontsize=12, weight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=9)

# Format y-axis with thousand separators
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:,.0f}'))

# Truncate long station names in x-axis
x_labels = [label.get_text() for label in ax.get_xticklabels()]
ax.set_xticklabels([label[:20] + '...' if len(label) > 20 else label for label in x_labels])

plt.legend(title='User Type', title_fontsize=12, fontsize=11, 
          loc='upper right', frameon=True, fancybox=True, shadow=True)

ax.grid(True, alpha=0.3, linestyle='-', axis='y')
ax.set_axisbelow(True)

plt.tight_layout()
plt.show()
```

```python
# What is the average trip distance for each rider type? (If we calculate using lat/lng) 

def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth's radius in kilometers
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c  # distance in kilometers

# Apply to dataset
df['trip_distance_km'] = haversine(df['start_lat'], df['start_lng'], df['end_lat'], df['end_lng'])

# Average distance by rider type
avg_distance = df.groupby('member_casual')['trip_distance_km'].mean().reset_index()
print(avg_distance)
```

```python
# Count rides per bike type for each user type
bike_popularity = (
    df.groupby(['member_casual', 'rideable_type'])['ride_id']
    .count()
    .reset_index(name='ride_count')
)

print(bike_popularity)
```

```python
sns.set_theme(style="whitegrid", font_scale=1.1)

plt.figure(figsize=(8, 5))
sns.barplot(data=bike_popularity, x='rideable_type', y='ride_count', hue='member_casual',
            palette=['#007bff', '#0040ff'],width=0.6)
plt.title('Popular Bike Type among Casual & Member', fontsize=14, weight='bold')
plt.ylabel('Ride Count')
plt.xlabel('Bike Type')
plt.tight_layout()

# Remove grid and borders
plt.grid(False)  
sns.despine(left=True, bottom=True)

plt.show()
```

```python
# How does bike usage vary seasonally for each group?

season_map = {
    12: 'Winter', 1: 'Winter', 2: 'Winter',
    3: 'Spring', 4: 'Spring', 5: 'Spring',
    6: 'Summer', 7: 'Summer', 8: 'Summer',
    9: 'Fall', 10: 'Fall', 11: 'Fall'
}

df['season'] = df['started_at'].dt.month.map(season_map)
```

```python
season_usage = (df.groupby(['member_casual', 'season'])['ride_id'].count().reset_index(name='ride_count'))
print(season_usage)
```

```python
sns.set_theme(style="whitegrid", font_scale=1.1)

plt.figure(figsize=(8, 5))
sns.barplot(data=season_usage, x='season', y='ride_count', hue='member_casual',
            palette=['#007bff', '#0040ff'], width=0.4)
plt.title('Seasonal Bike Usage by User Type', fontsize=14, weight='bold')
plt.ylabel('Ride Count')
plt.xlabel('Season')
plt.tight_layout()

# Remove grid and borders
plt.grid(False)  
sns.despine(left=True, bottom=True)

plt.show()
```
